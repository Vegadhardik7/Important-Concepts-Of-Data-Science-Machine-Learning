# Collection of Movie with Decision Tree

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("E:\\college_pics\\Rainbow6\\regression.csv",header=0)
```

```python
df.head()
```

```python
df.info()
```

### Handle Missing Values

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value  = df['Time_taken'].mean(), inplace = True)  #inplace = True store val permanently in df
```

```python
df.info()
```

### Convert Categorical Object Variable into Categorical Numerical Variable using Dummy Variable

```python
df.head()
```

#### There are 2 categorical object variable :
#### (i)3D_available : YES/NO
#### (ii)Genre : Thriller, Dramma, Comedy, Action

##### Number of Dummy variable should be : 1 - The number of category in each variable 

```python
df = pd.get_dummies(df,columns=["3D_available","Genre"],drop_first=True)
```

```python
df.head()
```

### X and y Split

```python
X = df.drop("Collection",axis=1)
type(X)
```

```python
X.head()
```

```python
X.shape
```

```python
y = df['Collection']
type(y)
```

```python
y.head()
```

```python
y.shape
```

### Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape # Use to train our model
```

```python
X_test.shape # Use to evaluate performace of our model
```

### Train Regression Tree

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(max_depth=3)     # max_depth < 5
```

```python
model.fit(X_train,y_train)
```

### Perdict values using trained model

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
y_test_pred
```

### Model Performance

```python
from sklearn.metrics import mean_squared_error,r2_score

# r2_score : R-Square value this is goodness of fit, lies between 0-1 , 1=perfect & 0=worst, 0.4-0.8=good
```

```python
mean_squared_error(y_test,y_test_pred)
```

```python
r2_score(y_train,y_train_pred)   # Based on model in which we trained it
```

```python
r2_score(y_test,y_test_pred)     # Completely new dataset
```

### Plotting Decision Tree

```python
from sklearn import tree
dot_data = tree.export_graphviz(model,out_file=None)
```

```python
from IPython.display import Image
```

```python
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

# Controlling Tree Growth

#### 1.) Maximun number of levels in tree

```python
model_1 = DecisionTreeRegressor(max_depth=3)
model_1.fit(X_train,y_train)
dot_data_1 = tree.export_graphviz(model_1,out_file=None,feature_names=X_train.columns,filled=True)
graph1 = pydotplus.graph_from_dot_data(dot_data_1)
Image(graph1.create_png())
```

#### 2.) Minimum Observation at internal node

```python
model_2 = DecisionTreeRegressor(min_samples_split=40)
model_2.fit(X_train,y_train)
dot_data_2 = tree.export_graphviz(model_2,out_file=None,feature_names=X_train.columns,filled=True)
graph2 = pydotplus.graph_from_dot_data(dot_data_2)
Image(graph2.create_png())
```

#### 3.) Minimum observation at leaf node 

```python
model_3 = DecisionTreeRegressor(min_samples_leaf=25, max_depth=4)
model_3.fit(X_train,y_train)
dot_data_3 = tree.export_graphviz(model_3,out_file=None,feature_names=X_train.columns,filled=True)
graph3 = pydotplus.graph_from_dot_data(dot_data_3)
Image(graph3.create_png())
```
