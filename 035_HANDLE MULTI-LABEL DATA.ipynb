```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'color':['Red','Green','Blue','Yellow','Orange','Pink','Black','Cyan','Grey','White'], 
                   'products':[['ball','car'], 
                               ['plant','vegetable'], 
                               ['book','ball'], 
                               ['car','bike'], 
                               ['plant', 'fruit'], 
                               ['dress','ball'], 
                               ['car', 'bike', 'ball'], 
                               ['car'], ['bike'], 
                               ['dress','book']]})

df.head(10)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

y = df['products']
y
```

```python
multi = MultiLabelBinarizer()
y = multi.fit_transform(df['products'])

y
```

```python
multi.classes_
```

```python
x = df['color']
x
```

```python
dataframe = pd.DataFrame(y,columns=multi.classes_)
dataframe
```

```python
dataframe.insert(0, column='color', value=x)
```
